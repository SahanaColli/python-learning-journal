```python
"""
FINAL CODE
"""
#define room 1
##repeat user interaction options (after each interaction is completed) until Path option is chosen
##print statement introducing game
##User input interaction options: talk to Trader, investigate Plant, leave to next room
##Trader- buy or sell interactions
###buy- display what Trader has to sell, let user input what item they want
####If statement to determine if user has enough money to buy item
###sell- displays what is in user's inventory, let user input what item they want to sell
####subtract sold item from inventory array, add item's worth in gold to user's gold value
##Plant item- abilty to add plant to inventory. Sellable
##Path- Move to the next room. use return to save any changed value, then end function
import random

#establish user values (by room1 for troubleshooting)
gold = 10
inventory = {"Rusty Sword": 2}
traderStock = {"Great Sword":10, "health potion":5   }
room = 1

def room1():
  room = 1
  rhubarb = True
  gold = 5
  userChoice = " "
  print("AT THE EDGE OF A TOWN AT THE TOP OF A STEEP PATH LIES A HORRIBLE LION. SHE TERRORIZES THE TOWNSPEOPLE BEFORE HIDING IN HER CAVE, GUARDED BY TWO WISE GUARDS. \nMANY ADVENTURERS HAVE TRIED TO SLAY HER TO TAKE THE TREASURE SHE GUARDS- NONE HAVE SUCCEEDED. WILL YOU?")
  print("Ahead of you lies a winding dirt path, with plants along either side. A trader looks up from their wares as you pass\nWHAT WILL YOU DO (enter one of the options below!)")
  while userChoice.lower() != "path":
    userChoice = input("Trader, plant, path: ") #repeats every loop until path is entered
    if userChoice.lower() == "trader":  #start of trader loop
      print("Trader: Hello there Adventurer! Looking for some supplies for your journey? Look no further! I've got the only Adventurer's gear in the area!")
      traderChoice = input("sell, buy, leave: ")
      if traderChoice == "sell":  #start of sell interaction
        sellChoice = input(f"What do you want to sell?\n{inventory}\t") #choose what to sell from inventory
        sellChoicePrice = traderStock.get(sellChoice) #retrieve price of chosen item from inventory
        sellable = False
        for item in inventory:  #check if chosen item is actually in inventory
          if item == sellChoice:
            sellable = True
        if sellable:
            traderStock[sellChoice] = [sellChoicePrice]   #adds sold item to trader stock
            traderStock[sellChoice].append(sellChoicePrice)   #adds sold item price to sold item in trader stock
            gold += sellChoicePrice   #add sold item price to user's gold
            del inventory[sellChoice] #remove the sold item from the user's inventory
            print("pleasure doing business with you")
            print(f"inventory: {inventory}")
        else:
          print("I don't have that")  #runs if chosen item isn't in inventory
      elif traderChoice == "buy":   #start of buy interaction
          buyChoice = input(f"what would you like to buy?\n{traderStock}\t")
          buyChoicePrice = traderStock.get(buyChoice)   #retrieve item user wants to buy's cost
          if buyChoicePrice > gold: #checks if user can actually afford item by comparing it's price to how much gold they have
            print("You can't afford that!")
          else:
            inventory[buyChoice] = [] #adds bought item to inventory
            inventory[buyChoice].append(buyChoicePrice) #adds bought item's price to inventory
            del traderStock[buyChoice]
            gold -= buyChoicePrice  #subtracts item price from user's gold value
            print("pleasure doing business with you")
            print(f"inventory: {inventory}")
    elif userChoice.lower() == "plant": #runs if user selects plant
      pickPlant = input("Its a rhubarb plant. pick some? (y/n) ")
      if (pickPlant.lower() == "y" or pickPlant.lower()== "yes") and rhubarb == True: #checks if user can pick/wants to pick rhubarb
          inventory["rhubarb"] = [] #adds rhubarb to inventory
          inventory["rhubarb"].append(5)  #adds rhubarb price to rhubarb in inventory
          print("got the plant")
          print(f"inventory: {inventory}")
          rhubarb = False #stops user from picking more rhubarb
      elif rhubarb == False:
        print("Nothings ripe. You leave it be.")
      else: #only prints if they say no before picking rhubarb
        print("You leave it be")
  room = 2
  return gold, inventory, room
#define room 2
##print statement introduction
##user input for interaction options- Guards, Door, Plant
##Plant item- abilty to add plant to inventory. Sellable
##Guards and Door both lead to riddle While Loop
##User is stuck in riddle while loop (While Door == Closed)until riddle is answered correctly
##Once Door == Open, repeat interaction options. Once door is chosen, end function

#establish room 2 values (by room2 definition for troubleshooting)
door = "closed"

def room2():
  room = 2
  global door #retrieve door state
  blackberry = True #set plant state
  attempts = 5  #set guesses amount for riddle
  print("At the end of the path you come across a cave opening covered by a large door. On either side of the door stand two guards. Blackberry bushes curl around the entrance.")
  userGuess = " "
  answer = "river"
  while room == 2:  #breaks if room value changes
    choice2 = input("GUARDS, DOOR, PLANT:\t") #user input for room 2 options
    if choice2.lower() == "guards" or choice2.lower() == "guard":
      print(f"Adventurer! To pass through the door to the Lion's den you must answer our riddle! If you get it wrong {attempts} times, we will strike you down!\n I have a mouth but never speak. \n I have a bed but never sleep.\n I run smoother than any rhyme.\n I love to fall but cannot climb.")
      while userGuess.lower != answer:
        userGuess = (input("What am I? "))
        if userGuess.lower() == answer: #breaks riddle loop if user guesses correctly
          print("Correct! you can move on")
          door = "open"
          break
        elif userGuess == "leave":  #exit riddle
          userGuess = " "
        else:
          print("Wrong! try again")
          attempts -= 1 #removes attempt but continues loop as long as their is attempts remaining
        if attempts == 0:
          print("Too many guesses! The guards kill you")
          room = 0
          break
    elif choice2.lower() == "door":   #moves to room 3 if open
      if door == "open":
        print("You push open the old door.")
        room = 3
      else: #only prints if riddle has been answered correctly
        print("the door guards block the door. Maybe you should see why?")
    elif choice2.lower() == "plant":  #same as rhubarb code
      pickPlant = input("Its a blackberry bush. pick some? (y/n) ")
      if (pickPlant.lower() == "y" or pickPlant.lower()== "yes") and blackberry == True:  #can only pick blackberry once
          inventory["blackberry"] = []
          inventory["blackberry"].append(5)
          print("got the plant")
          print(f"inventory: {inventory}")
          blackberry = False
      elif blackberry == False:
          print("Nothings ripe. You leave it be.")
      else:
          print("You leave it be")
  return room #updates room value


#room 3 values (by room3 definition for troubleshooting)
userHP = 12
lion = "alive"
lionHP = 10
def room3():
  global room #retrieve important values
  global userHP
  global lion
  global lionHP
  global door
  lionDrop = {"lion hide":2, "gold":3}  #dictionary for prizes if user beats the lion
  ##Only option attack lion, trap user in While loop until either lion or user hp reaches 0
  ##switch between lionTurn(random attack choice) and userTurn(user is given input options on how to attack, changing damage dealt)
  print("As you enter the cave, you see the lion rise to her full height. She bares her teeth")
  print("The lion lunges! The battle is on!")
  while lionHP > 0 and userHP > 0 and lion == "alive":
  #user's turn
    defended = False
    userAttack = input("what will you do? ATTACK, DEFEND, POTION, HELP: ")  #fight choices
    if userAttack.lower() == "attack":  #lowers lion hp
      lionHP -= random.randint(1,5)
    elif userAttack.lower() == "defend":  #stops lion attack
      defended = True
    elif userAttack.lower() == "potion":  #heals user 5 hp if user has potion
      usePotion = False
      for item in inventory:
        if item == "health potion": #checks if user has potion
          usePotion = True
      if usePotion:
        userHP += 5
        print("You drank a potion")
        del inventory["health potion"]  #removes potion from user inventory
      else:
        print("You don't have a health potion")
    else:
      print("ATTACK: lowers the lion's HP. Once the lion HP reaches 0, you win!\nDEFEND: block against the lion's attack\nPOTION: use a health potion to heal yourself")
    #lion's turn
    if lionHP > 0:  #checks if lion is dead
      if userAttack.lower() == "defend":  #other half of user defend command- effectively stops lion's turn
        print("The lion ATTACKs you! but you manage to DEFEND against it!")
      elif userAttack == "help":  #stops lion from attacking if user was asking for help- doesn't include mispellings
        continue
      else:                            #lion's only move- lowers user hp
        print("The lion ATTACKs you!")
        lionAttack = random.randint(0,4)
        userHP -= lionAttack
      print(f"your health: {userHP}\nLion's health: {lionHP}")
  ###lion reaches 0 first, user wins. release from room 3 function and continue to next room
  ###user reaches 0 first, user loses. End code early
  if lionHP <= 0:
    print("The lion is dead. You win!")
    print(f"Got: {lionDrop}")
    lion = "dead"
    inventory.update(lionDrop)
  elif userHP <= 0:                         #stops program if user loses
    print("the lion kills you! You Lose!")
    room = 0    #stops while loop
    lion = "alive"       #stops finalChoice options
  if lion == "dead":  #only runs if user hasn't died
    finalChoice = input("Past the lion's corpse a small corridor is made clear. Go through it? (y/n) ")
    if finalChoice.lower() == "yes" or finalChoice.lower() == "y":  #triggers win ending
      print("You squeeze your way down the narrow corridor...")
      room = 4      #triggers to new room
    elif finalChoice.lower() == "no" or finalChoice.lower() == "n":
        print("Thats enough Adventure for the day. You leave the cave and head home...\n\nTHANKS FOR PLAYING!!")
        room = 0
  return room

riches = {"gold":500, "Jewels": 64, "nebulous treasure": 12}  #vague treasure in room 4

#actual running code
while room != 0: #keeps room running as long as the user does not fail (only time room = 0)
  if room == 1: #runs the room 1 code
    gold, inventory, room = room1() #runs room 1 and returns important values
    print(f"\ninventory:{inventory}\ngold: {gold}") #tells user inventory and gold values
  elif room == 2:
    room = room2()  #runs room 2 and returns important values (just room)
    print(f"\ninventory:{inventory}\ngold: {gold}") #tells user inventory and gold values
  elif room == 3: #runs room 2 and returns important values (just room)
    room = room3()
  elif room == 4: #runs winner/treasure room- didn't bother making a function cause its like 4 lines
    print("As you venture through the cave the corridor opens up into a large room.. Full of golden jewels and treasure as far as the eye can see!")
    print(f"got {riches}")
    inventory.update(riches)
    print(f"your final inventory: {inventory}\n\n THANKS FOR PLAYING!!!")
    room = 0
```
